```python
def seekmonth(x):
    if str(x)[:3] == "Jan":
        return '1'
    elif str(x)[:3] == "Feb":
        return '2'
    elif str(x)[:3] == "Mar":
        return '3'
    elif str(x)[:3] == "Apr":
        return '4'
    elif str(x)[:3] == "May":
        return '5'
    elif str(x)[:3] == "Jun":
        return '6'
    elif str(x)[:3] == "Jul":
        return '7'
    elif str(x)[:3] == "Aug":
        return '8'
    elif str(x)[:3] == "Sep":
        return '9'
    elif str(x)[:3] == "Oct":
        return '10'
    elif str(x)[:3] == "Nov":
        return '11'
    elif str(x)[:3] == "Dec":
        return '12'
    elif str(x)[5:8] == "Jan":
        return '1'
    elif str(x)[5:8] == "Feb":
        return '2'
    elif str(x)[5:8] == "Mar":
        return '3'
    elif str(x)[5:8] == "Apr":
        return '4'
    elif str(x)[5:8] == "May":
        return '5'
    elif str(x)[5:8] == "Jun":
        return '6'
    elif str(x)[5:8] == "Jul":
        return '7'
    elif str(x)[5:8] == "Aug":
        return '8'
    elif str(x)[5:8] == "Sep":
        return '9'
    elif str(x)[5:8] == "Oct":
        return '10'
    elif str(x)[5:8] == "Nov":
        return '11'
    elif str(x)[5:8] == "Dec":
        return '12'
    else:
        return ""
```

```python
def seekday(x):
    if str(x)[5:7] in str(np.arange(32)):
        return str(x)[5:7]
    elif str(x)[5] in str(np.arange(32)):
        return str(x)[5]
    else:
        if str(x)[10:12] in str(np.arange(32)):
            return str(x)[10:12]
```

```python
def seekyear(x):
    yr = ['2012', '2013', '2014', '2015', '2016', '2017']
    if str(x)[9:13] in yr:
        return str(x)[9:13]
    elif str(x)[8:12] in yr:
        return str(x)[8:12]
    else:
        return '2017'
```

```python
seekalpha['Month'] = list(map(lambda x: seekmonth(x), seekalpha.Date))
```

```python
seekalpha['Day'] = list(map(lambda x: seekday(x), seekalpha.Date))
```
